```python
import pandas as pd

def read_csv(file_path):
    """Read a CSV file and return a DataFrame."""
    try:
        df = pd.read_csv(file_path, sep=';')
        print(f"Successfully read file: {file_path}")
        return df
    except Exception as e:
        print(f"Error reading file {file_path}: {e}")
        raise

def merge_dataframes(df1, df2, join_column, how="inner"):
    """Merge two DataFrames on a common column with prefixes."""
    try:
        merged_df = pd.merge(df1.add_prefix("1_"), df2.add_prefix("2_"), left_on=f"1_{join_column}", right_on=f"2_{join_column}", how=how)
        merged_df = merged_df.rename(columns={f"1_{join_column}": join_column}).drop(columns=[f"2_{join_column}"])
        print(f"Successfully merged DataFrames on column: {join_column}")
        return merged_df
    except Exception as e:
        print(f"Error merging DataFrames: {e}")
        raise

def write_to_csv(df, output_path):
    """Write a DataFrame to a CSV file."""
    try:
        df.to_csv(output_path, index=False, sep=";")
        print(f"Successfully wrote merged data to file: {output_path}")
    except Exception as e:
        print(f"Error writing to file {output_path}: {e}")
        raise

def main():
    # Input paths
    input_file_1 = "/content/drive/MyDrive/Colab Notebooks/TopEditor/data/Textenrichment/enrichement_lugares_JEAN_TOPEDITOR_Fourniture_IdRef_102024_lugares.csv"
    input_file_2 = "/content/drive/MyDrive/Colab Notebooks/TopEditor/data/Textenrichment/enrichement_metiers_SARA_TOPEDITOR_Fourniture_IdRef_102024_avecOpentheso.csv"
    output_file = "/content/drive/MyDrive/Colab Notebooks/TopEditor/data/Textenrichment/enrichement_TOPEDITOR_Fourniture_IdRef_merged_12122024_personnes_juives.csv"

    # Common column to join on
    join_column = "IDREF"  # Change this to the name of your column

    # Read the input files
    df1 = read_csv(input_file_1)
    print(f"Number of rows in {input_file_1}: {len(df1)}")
    df2 = read_csv(input_file_2)
    print(f"Number of rows in {input_file_2}: {len(df2)}")

    # Merge the data
    merged_df = merge_dataframes(df1, df2, join_column)
    print(f"Number of rows in merged DataFrame: {len(merged_df)}")

    # Write the merged data to an output file
    write_to_csv(merged_df, output_file)

if __name__ == "__main__":
    main()
```
