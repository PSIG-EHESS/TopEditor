````python
directory_json = "/content/drive/MyDrive/Colab Notebooks/TopEditor/data/tagtog/annotations/sara_12_10_2023/"
directory_tr = "/content/drive/MyDrive/Colab Notebooks/TopEditor/data/tagtog/extraits_textes/"
directory_json_labels = "/content/drive/MyDrive/Colab Notebooks/TopEditor/data/tagtog/json_labelstudio_conv/"

class_lists = ['PER', "Oficio", "Religion", "TipoContratoLT", "nombre_collacion",
                              "nombre_calle", "ORG", "modo_pago", "LOC",
                              "Titulo_role_statut", "rol_pers_familia", "piezas_casas", "nombre_CTMB",
                              "Alias_persona", "estado_casa", "Fecha_periodos_eventos", "DEPREC_fiador_testigo",
                              "medidas", "relations_spatiales", "tipo_edificio", "ExpresionesJuridicas",
                              "TipoContratoCT", "TipoContratoPerpetua", "elemento_arquitectura", "expresion_medidas",
                              "Obras", "Adefinir30", "Adefinir31"]

def get_class_id_inv(label):
  choices = {'e_1': 'PER', 'e_2' : 'Oficio', 'e_3' : 'adefinir', 'e_4' : 'Religion', 'e_5' : 'adefinir2', 'e_6' : 'TipoContratoLT',
             'e_7' : 'nombre_collacion', 'e_8' : 'nombre_calle', 'e_9' : 'ORG', 'e_10' : 'modo_pago', 'e_11' : 'LOC',
             'e_12' : 'Titulo_role_statut', 'e_13' : 'rol_pers_familia', 'e_14' : 'piezas_casas', 'e_15' : 'nombre_CTMB',
             'e_16' : 'Alias_persona', 'e_17' : 'estado_casa', 'e_19' : 'Fecha_periodos_eventos', 'e_20' : 'DEPREC_fiador_testigo',
             'e_21': 'medidas', 'e_22' : 'relations_spatiales', 'e_23' : 'tipo_edificio', 'e_24' : 'ExpresionesJuridicas',
             'e_25' : 'TipoContratoCT', 'e_26' : 'TipoContratoPerpetua', 'e_27' : 'elemento_arquitectura', 'e_28' : 'expresion_medidas',
             'e_29' : 'Obras', 'e_30' : 'Adefinir30', 'e_31' : 'Adefinir30' }
  return choices.get(label, None)
````

````python
import json, glob

list_training_data = []

for filename in glob.glob(directory_json+'/*.json'):
  filen = filename.split("/")[-1]
  print(filen)
  training_data =  {'filename' : filen, 'classes' : class_lists, 'annotations' : []}
  with open(filename, "r", encoding="utf_8") as f:
    data = f.read()
    js_corrected_training_data = json.loads(data)
    temp_dict = {}
    print(directory_tr + filen.strip().replace(".ann.json","").replace("_","-"))
    fin = open(directory_tr + filen.strip().replace(".ann.json","").replace("_","-"), "r")
    temp_dict['text'] = fin.read()
    fin.close()
    temp_dict['entities'] = []
    for example in js_corrected_training_data['entities']:
      #print(example)
      start = example['offsets'][0]['start']
      end = int(example['offsets'][0]['start']) + len(example['offsets'][0]['text'])
      label = get_class_id_inv(example['classId'])
      anntxt = example['offsets'][0]['text']
      temp_dict['entities'].append((start, end, label, anntxt))
      #print(temp_dict['entities'])
    training_data['annotations'].append(temp_dict)
    list_training_data.append(training_data)
````

````python
import json

for example in list_training_data:
  count = 0
  results = []
  for ent in example['annotations'][0]['entities']:
    results.append({
              "id": "ann_"+str(count),
              "from_name": "label",
              "to_name": "text",
              "type": "labels",
              "value": {
                "start": ent[0],
                "end": ent[1],
                "score": 1,
                "text": ent[3],
                "labels": [
                  ent[2]
                ]
              }
            })
    count = count + 1

  ann = [
      {
      "data": {
        "text": example['annotations'][0]['text']
      },
      "predictions": [
        {
          "model_version": "one",
          "score": 1,
          "result" : results
        }
      ]
    }
  ]
  json_object = json.dumps(ann, ensure_ascii=False, indent=4)
  with open(directory_json_labels+example["filename"]+".json", "w", encoding="utf-8") as outfile:
    outfile.write(json_object)
````

Decoupage du CONLL entier en morceaux pour travailler de manière incrementale

````python
conll_name = "/content/drive/MyDrive/Colab Notebooks/TopEditor/data/conll/project-13-at-2024-02-19-16-49-3186e026_v16042024.conll"
folder_out = "/content/drive/MyDrive/Colab Notebooks/TopEditor/data/conll/"

def split_conll_file(input_file, output_prefix, num_parts):
    # Open the input file for reading
    with open(input_file, 'r', encoding='utf-8') as f:
        lines = f.readlines()

    # Calculate the size of each part
    total_lines = len(lines)
    print(total_lines)
    lines_per_part = total_lines // num_parts
    print(lines_per_part)
    remaining_lines = total_lines % num_parts  # Calculate remaining lines
    print(remaining_lines)

    # Split the file into parts and write each part to a separate output file
    start_idx = 0
    for i in range(num_parts):
        if i < remaining_lines:
            end_idx = start_idx + lines_per_part + 1
        else:
            end_idx = start_idx + lines_per_part + remaining_lines

        # Create the output file name for this part
        output_file = f"{output_prefix}_{i + 1}.conll"

        # Write the lines to the output file
        print(folder_out + output_file)
        with open(folder_out + output_file, 'w', encoding='utf-8') as out_f:
          print(len(lines[start_idx:end_idx]))
          out_f.writelines(lines[start_idx:end_idx])

        # Update start index for the next part
        start_idx = end_idx

    print(f"Output files created: {num_parts} parts")

# Utilisation de la fonction avec un fichier d'entrée "input.conll" et 10 parties de sortie
split_conll_file(conll_name, "annot_data_topeditor", 20)
````

Découper chaque fichier en 3 parties pour faciliter la correction manuelle

````python
def split_file_in_three(input_file, output_prefix):
    # Lire le contenu du fichier d'entrée
    with open(input_file, 'r') as file:
        lines = file.readlines()

    # Déterminer le nombre total de lignes
    total_lines = len(lines)

    # Calculer la taille de chaque partie
    part_size = total_lines // 3

    # Déterminer les indices de début et de fin pour chaque partie
    indices = [(0, part_size), (part_size, 2 * part_size), (2 * part_size, total_lines)]

    # Créer les fichiers de sortie pour chaque partie
    fold = "/content/drive/MyDrive/Colab Notebooks/TopEditor/data/conll/"
    for i, (start, end) in enumerate(indices):
        output_file = f"{output_prefix}_part_{i+1}.txt"
        with open(fold + output_file, 'w') as out_file:
            out_file.writelines(lines[start:end])

    print("Fichier découpé en 3 parties avec succès.")

# Exemple d'utilisation
split_file_in_three('/content/drive/MyDrive/Colab Notebooks/TopEditor/data/conll/annot_data_topeditor_15.conll', 'annot_data_topeditor_15')
````

Decouper en 30 parts

````python
import os

def split_files_in_thirty(input_folder, output_folder):
    # Vérifier si le dossier de sortie existe, sinon le créer
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    # Parcourir tous les fichiers du dossier d'entrée
    for filename in os.listdir(input_folder):
        # Travailler seulement avec les fichiers texte
        if filename.endswith('.conll'):
            input_file = os.path.join(input_folder, filename)
            print(input_file)
            # Lire le contenu du fichier d'entrée
            with open(input_file, 'r') as file:
                lines = file.readlines()

            # Déterminer le nombre total de lignes
            total_lines = len(lines)
            print(total_lines)

            # Calculer la taille de chaque partie
            part_size = total_lines // 30
            print(part_size)
            # Déterminer les indices de début et de fin pour chaque partie
            indices = [(i * part_size, (i + 1) * part_size) for i in range(29)]
            indices.append((29 * part_size, total_lines))  # La dernière partie prend le reste des lignes

            # Nommer les fichiers de sortie et écrire les parties
            for i, (start, end) in enumerate(indices):
                output_file = os.path.join(output_folder, f"{os.path.splitext(filename)[0]}_part_{i+1}.conll")
                with open(output_file, 'w') as out_file:
                    out_file.writelines(lines[start:end])

            print(f"Fichier {filename} découpé en 30 parties avec succès.")

# Exemple d'utilisation
split_files_in_thirty('/content/drive/MyDrive/Colab Notebooks/TopEditor/data/conll/30parts/', '/content/drive/MyDrive/Colab Notebooks/TopEditor/data/conll/30parts/')

````

````python
import os

def delete_part_files(folder):
    # Parcourir tous les fichiers du dossier
    for filename in os.listdir(folder):
        # Vérifier si "part" est dans le nom du fichier
        if "part" in filename:
            file_path = os.path.join(folder, filename)
            if os.path.isfile(file_path):  # Assurer que c'est bien un fichier
                os.remove(file_path)  # Supprimer le fichier
                print(f"Fichier supprimé: {filename}")

# Exemple d'utilisation
folder = '/content/drive/MyDrive/Colab Notebooks/TopEditor/data/conll/20parts'  # Remplacez par le chemin de votre dossier
delete_part_files(folder)
````
