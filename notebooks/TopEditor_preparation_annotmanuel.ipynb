```python
import docx2txt, glob, random

dossier = "/content/drive/MyDrive/EHESS/FNSOTopEditor/Documentos estilados WORD"
dossier_out = "/content/drive/MyDrive/EHESS/FNSOTopEditor/Codificación_textos/Anotacion_textos/corpus_manual"


for filename in glob.glob(dossier+'/*.docx'):
  filen = filename.split("/")[-1]
  print(filen)
  text = docx2txt.process(dossier+"/"+filen)
  content = text.split("\n")
  file_len = len(content) #nombre de phrases par fichier
  taill_un_tiers = int(file_len / 3)
  print(taill_un_tiers)
  
  f = open(dossier_out+"/"+filen+".txt", "w")
  for item in content[2:int(taill_un_tiers/2)]:
    f.write(item+"\n")
  f.write("\n\n")
  
  #for item in content[int(file_len/2):int(file_len/2) + taill_un_tiers]:
  #  f.write(item+"\n")
  #f.write("\n\n")
  
  for item in content[-int(taill_un_tiers/2):]:
    f.write(item+"\n")
  f.write("\n\n")
  
  f.close()

```
